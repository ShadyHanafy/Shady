```python
df['host_is_superhost'].isnull().value_counts()
```

```python
#Convert the chosen cateogory columns using get dummies to be able to use it in our model

df_cat = pd.concat([df.drop('host_is_superhost', axis=1), pd.get_dummies(df['host_is_superhost'], prefix='host_is_superhost', prefix_sep='_', drop_first=True)], axis=1)
df_catg = pd.concat([df_cat.drop('cancellation_policy', axis=1), pd.get_dummies(df_cat['cancellation_policy'], prefix='cancellation_policy', prefix_sep='_', drop_first=False)], axis=1)
df_categ = pd.concat([df_catg.drop('instant_bookable', axis=1), pd.get_dummies(df_catg['instant_bookable'], prefix='instant_bookable', prefix_sep='_', drop_first=True)], axis=1)
```

```python
df_categ.head()
```

```python
#Check the hospitality index using groupby and removing the empty values if exist
hospitality=df_categ.groupby(['neighbourhood_group_cleansed']).mean()['host_is_superhost_t'].sort_values().dropna()
(hospitality).plot(kind="bar");
plt.title("The hospitality index");
```

```python
#Check the accomodation capacity using groupby and removing the empty values if exist
accomodation=df_categ.groupby(['neighbourhood_group_cleansed']).mean()['accommodates'].sort_values().dropna()
(accomodation).plot(kind="bar");
plt.title("The accomodation capacity");
```

```python
#Check the number of bathrooms using groupby and removing the empty values if exist
num_bathrooms=df_categ.groupby(['neighbourhood_group_cleansed']).mean()['bathrooms'].sort_values().dropna()
(num_bathrooms).plot(kind="bar");
plt.title("The Number of Bathrooms");
```

```python
#Check the number of bedrooms using groupby and removing the empty values if exist
num_bedrooms=df_categ.groupby(['neighbourhood_group_cleansed']).mean()['bedrooms'].sort_values().dropna()
(num_bedrooms).plot(kind="bar");
plt.title("The Number of Bedrooms");
```

```python
#Check the number of rooms of each neighborhood property type
roomtype=df_categ.groupby('neighbourhood_group_cleansed').room_type.value_counts()
plt.figure(figsize=(15,4))
roomtype.plot(kind = 'bar', title="Rooms")
plt.xticks(rotation=90)
plt.show()
```

```python
#Check how many guests can be included
Guests=df_categ.groupby(['neighbourhood_group_cleansed']).mean()['guests_included'].sort_values().dropna()
(Guests).plot(kind="bar");
plt.title("The guests includance");
```
